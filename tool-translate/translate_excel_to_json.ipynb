```python
import pandas as pd
import json
import re
import os

# Đọc tệp Excel
file_path = './Translation_BacHa_Weather.xlsx'  # thay thế bằng đường dẫn tệp thực tế
df = pd.read_excel(file_path)

# Tạo thư mục 'translates' nếu chưa tồn tại
output_dir = '../src/locales/translations'
os.makedirs(output_dir, exist_ok=True)

# Lặp qua các cột bắt đầu từ cột thứ ba (bỏ qua cột 'Key' và 'Default')
for col in df.columns[2:]:
    # Tạo từ điển để lưu bản dịch cho từng ngôn ngữ
    translations = {}
    
    for index, row in df.iterrows():
        # Sử dụng cột 'Key' làm khóa và cột ngôn ngữ tương ứng làm giá trị
        translations[row['Key']] = row[col]
    
    # Chuyển đổi từ điển thành JSON
    json_output = json.dumps(translations, ensure_ascii=False, indent=4)
    
    # Trích xuất mã ngôn ngữ từ tên cột
    match = re.search(r'\(([\w-]+)\)', col)
    if match:
        lang_code = match.group(1)
    else:
        lang_code = col  # nếu không tìm thấy mã ngôn ngữ, sử dụng tên cột gốc
    
    # Tạo tên tệp JSON với mã ngôn ngữ trong thư mục 'translates'
    output_file = os.path.join(output_dir, f'{lang_code}.json')
    
    # Ghi dữ liệu JSON vào tệp
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(json_output)
    
#     print(f'Đã lưu bản dịch cho {col} vào tệp {output_file}')
#     print(f'import {lang_code}Value from "./translations/{lang_code}.json";')
    print(f'{lang_code}: {lang_code}Value,')
    
    
```
